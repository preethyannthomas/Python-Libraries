**Python Data Slicing**

1. Selecting data according to some conditions

```python
display(df.loc[(df.Team == 'Utah Jazz') & (df.Age>25)])
```

2. Selecting a range of rows from DataFrame

```python
df2 = df.loc[448:451]
df2
```

```python
df.loc[(df.Name=='Gordon Hayward'),['Salary']] = 190000 # updating columns
```

```python
display(df2)
```

1. Selecting rows using integer indices (iloc)

```python
df3 = df.iloc[448:451]
df3
```

```python
display(df.iloc[448:451,0:4]) # row, col
```

```python
df4 = df3.drop(index=[450],axis=0)
display(df4)
```

```python
df5 = df3.drop('Number',axis=1 ) # axis to represent col
display(df5)
```

```python
display(df4.drop(columns='Number'))
print(df4.drop(columns='Number'))
```

**Lambda Functions**

```python
def double(x):
  return x*2
double(10)
```

Anonymous functions are defined using lambda keyword.

Syntax:       lambda arguments: expression

```python
double = lambda x:x*2
print(double(10))
```

```python
mean_meta = df_movie['Metascore'].mean()
mean_meta
```

```python
meta = df_movie['Metascore'].apply(lambda x: x if x == x else mean_meta)
meta
```
