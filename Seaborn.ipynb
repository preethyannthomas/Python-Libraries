**SEABORN**

Data visualization library based on Matplotlib and provides a high level interface for drawing attractive and informative statistical graphics.

It uses fewer syntax and has stunning default thems

```python
# !PIP INSTALL SEABORN
import seaborn as sns
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.relplot(x='total_bill', y='tip',data=tips) # by default plots scatterplot
```

**Hue, Style and Size**

In seaborn, the hue parameter determines which column in the data frame should be used for colour encoding.

Seaborn has five built-in themes to style its plots: darkgrid , whitegrid , dark , white , and ticks . Seaborn defaults to using the darkgrid theme for its plots, but you can change this styling to better suit your presentation needs. To use any of the preset themes pass the name of it to sns.

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue = 'time')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue = 'smoker', style = 'time')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue = 'smoker', style = 'time', size = 'size')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, style = 'size')
```

```python
tips['size'].value_counts()
```

**Line Plot**

```python
sns.set(font_scale=2)
sns.relplot(x='total_bill', y='tip', kind = 'line', data=tips, height = 9, aspect = 4)
```

```python
sns.set(font_scale=2)
sns.relplot(x='total_bill', y='tip', kind = 'line', data=tips, height = 3, aspect = 4)
```

```python
sns.relplot(x='total_bill', y='tip', kind = 'line', hue = 'smoker', data=tips, height = 9, aspect = 4)
plt.legend(loc='right')
```

```python
sns.set(font_scale=0.75)
sns.relplot(x='total_bill', y='tip', kind = 'line', hue = 'smoker', data=tips, height = 9, aspect = 4)
plt.legend(loc='right')
```

**Subplot**

```python
tips.head()
```

```python
sns.set(font_scale=1.5)
sns.relplot(x='total_bill', y='tip', col = 'smoker', hue = 'smoker', data=tips)
```

```python
sns.relplot(x='total_bill', y='tip', col = 'size', hue = 'smoker', data=tips)
```

```python
sns.relplot(x='total_bill', y='tip', col = 'time', hue = 'smoker', data=tips, row = 'size')
```

```python
sns.relplot(x='total_bill', y='tip', col = 'size', hue = 'smoker', data=tips, col_wrap = 3)
```

**Categorical Data Ploting**

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, jitter = False) # to reduce spread like above fig use jitter
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'swarm', hue = 'size')
```

```python
sns.catplot(x = 'smoker', y = 'tip', data = tips, order = ['No','Yes'])
```

```python
sns.catplot(x='day', y='total_bill', kind='box', data=tips, hue = 'sex')
```

**Bar Plot**

```python
sns.catplot(x='day', y='total_bill', kind='bar', data=tips, hue = 'sex')
```

**Point Plot**

```python
sns.catplot(x='day', y='total_bill', kind='point', data=tips, hue = 'smoker')
```

**Pair Plot**

```python
sns.pairplot(tips)
```

**Regression Plot**
